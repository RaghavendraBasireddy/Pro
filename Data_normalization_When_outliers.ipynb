```python
How do we know there is an outliers in your Data,
 An outlier is a data point which significantly differs with other data points. Having outliers in data causes abnormalities in Predictions done by the machine learning model.
There several ways to detect outliers

One of them is Box Plot method, box plot whiskers detects the outliers which are unusually low or high values in a variable or data set using IQR(Interquartile Value ) value. The general value for interquartile value is 1.5 any points greater than this value are treated as potential outliers.

 

  
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('C:\\Users\\134033\\Desktop\\Python Mate\\abalonedata.csv')
df.columns = ('sex','Length','Diameter','Height','Whole Weigth','Shcked Weight','Visecera Weight','Shall Weight','Rings')
df.head()
```
