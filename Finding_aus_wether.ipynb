```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('E:\\Downloads\\weather-dataset-rattle-package\\weatherAUS.csv')

df
```

```python
df = df.drop(['Date','Location','Evaporation','Sunshine','Cloud9am','Cloud3pm','RISK_MM'],axis=1)
```

```python
df['RainToday']=df['RainToday'].replace({'No':0,'Yes':1})

df['RainTomorrow'] = df['RainTomorrow'].replace({'No':0,'Yes':1})
```

```python
df['WindGustDir'] = pd.get_dummies(df['WindGustDir'])

df['WindGustSpeed'] = pd.get_dummies(df['WindGustSpeed'])

df['WindDir9am'] = pd.get_dummies(df['WindDir9am'])

df['WindDir3pm'] = pd.get_dummies(df['WindDir3pm'])
```

```python
df = df.dropna(how='any')
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lben = LabelEncoder()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.ensemble  import BaggingClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='entropy',max_depth= 5,min_samples_split=2)

dtree_gini = DecisionTreeClassifier(criterion='gini',max_depth=7,min_samples_split=2)
                                   
```

```python
Bg_classifier_dtree = BaggingClassifier(base_estimator = dtree,n_estimators=10,max_features=10)
```

```python
Bg_classifier_dtree.fit(x_train,y_train)
```

```python
y_pred = Bg_classifier_dtree.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
Bg_classifier_gini = BaggingClassifier(base_estimator= dtree_gini)
```

```python
Bg_classifier_gini.fit(x_train,y_train)
```

```python
y_pred = Bg_classifier_gini.predict(x_test)
```

```python
cm_gini = confusion_matrix(y_test,y_pred)
print(cm_gini)
```
